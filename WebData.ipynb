# Offensive Stats - Goals per Minutes Played Chart And Graph

```python
# Create a new DateFrame for EPL_Players_df
player_goals_df = EPL_Players_df[["Player", "Min", "Goals"]].head(20)
player_goals_df = player_goals_df.set_index('Player')
player_goals_df
```

```python
# Create a SCATTER Graph for MINUTES PLAYED VS GOALS SCORED (Top 20 Players by Goals Scored)
# player_goals_df = player_goals_df.sort_values("Min", ascending=True)
player_goals_df = player_goals_df.reset_index()
x_axis = player_goals_df['Min']
y_axis = player_goals_df["Goals"]

plt.figure(figsize=(16, 10))
plt.grid()
plt.scatter(x_axis, y_axis)
plt.xlabel('Minutes Played')
plt.ylabel('Goals')
plt.xticks(rotation="vertical")

labels = player_goals_df['Player'].tolist()

for x, y, label in zip(x_axis, y_axis, labels):
    plt.annotate(label,
                xy=(x, y),
                rotation=30,
#                 textcoords="offset points", # how to position the text
#                 xytext=(0,10), # distance from text to points (x,y)
#                 ha='center'
                )

    
plt.show()
```

# Offensive Stats - Assists per Minutes Played Chart And Graph

```python
# Create a new DateFrame for EPL_Players_df
player_ast_df = EPL_Players_df[["Player", "Min", "Ast"]]
player_ast_df = player_ast_df.sort_values("Ast", ascending=False).head(20)
player_ast_df = player_ast_df.set_index('Player')
player_ast_df
```

```python
# Create a SCATTER Graph for MINUTES PLAYED VS ASSISTS (Top 20 Players by Asssits)
# player_ast_df = player_ast_df.sort_values("Ast", ascending=True)
player_ast_df = player_ast_df.reset_index()
x_axis = player_ast_df['Min']
y_axis = player_ast_df["Ast"]

plt.figure(figsize=(16, 10))
plt.grid()
plt.scatter(x_axis, y_axis)
plt.xlabel('Minutes Played')
plt.ylabel('Assists')
plt.xticks(rotation="vertical")

labels = player_ast_df['Player'].tolist()
for x, y, label in zip(x_axis, y_axis, labels):
    plt.annotate(label,
                xy=(x, y),
                rotation=45)

plt.show()
```
